# Feature Distribution Analysis
This notebook inspects feature distributions before any model training begins. It follows a structured process for numerical and categorical evaluation, including target-based comparisons and transformation recommendations.

## 1. Import Libraries and Load Dataset

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.eda import (
    DATA_PATH,
    load_data,
    generate_sample_dataset,
    describe_numeric_features,
    plot_numeric_distributions,
    identify_numeric_outliers,
    inspect_categorical_features,
    compare_numeric_by_target,
    plot_compare_by_target,
)

sns.set_theme(style="whitegrid")

try:
    df = load_data()
    print(f"Loaded dataset from {DATA_PATH}")
except FileNotFoundError:
    print(f"Dataset not found at {DATA_PATH}. Generating a synthetic sample dataset for analysis.")
    df = generate_sample_dataset()
    print(f"Generated sample dataset at {DATA_PATH}")

print("
Dataset shape:", df.shape)
df.head(5)
```

## 2. Inspect Numerical Features

```python
numeric_summary = describe_numeric_features(df)
print(numeric_summary)

numeric_summary
```

## 3. Plot Numerical Distributions

```python
plots = plot_numeric_distributions(df)
print("Saved plots:")
for name, path in plots.items():
    print(f"- {name}: {path}")

from IPython.display import Image, display
for path in plots.values():
    display(Image(filename=path))
```

## 4. Compute Skewness and Identify Outliers

```python
outliers = identify_numeric_outliers(df)
print(outliers)

outliers
```

## 5. Inspect Categorical Features

```python
categorical_inspection = inspect_categorical_features(df)
for col, info in categorical_inspection.items():
    print(f"\nColumn: {col}")
    print("Value counts:")
    print(info["counts"].head(20))
    print("Rare categories (<1%):")
    print(info["rare_categories"])
    print(f"Unique values: {info['unique_values']}")
    print(f"Rare category percentage: {info['rare_percentage'] * 100:.2f}%")
```

## 6. Compare Features by Target

```python
target_cols = ["age", "length_of_stay"]
comparison = compare_numeric_by_target(df, target_cols)
print(comparison)

comparison_plot = plot_compare_by_target(df, target_cols)
print(f"Saved target comparison plot: {comparison_plot}")
from IPython.display import Image, display
display(Image(filename=comparison_plot))
```

## 7. Summarize Recommended Transformations
This section documents the main observations from the feature distribution inspection and identifies transformations that should be considered before training.

- Numerical features with strong positive skew may require log or power transformations.
- Features with heavy outliers should be clipped or winsorized before model fitting.
- Categorical variables with rare levels should be grouped into an `Other` category if they are too sparse.
- No transformations are fit on the entire dataset; this notebook only inspects raw data.
